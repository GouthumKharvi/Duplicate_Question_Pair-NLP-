# Create New Dataframe

```python
ques_df = new_df[['question1','question2']]
ques_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# Merge the text data from two columns into a single list of strings
questions = list(ques_df['question1']) + list(ques_df['question2'])

# Initialize the CountVectorizer to convert text into a matrix of token counts
# max_features=3000 limits the number of features (words) to the top 3000 most frequent words
cv = CountVectorizer(max_features=3000)
q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
temp_df1 = pd.DataFrame(q1_arr, index= ques_df.index)
temp_df2 = pd.DataFrame(q2_arr, index= ques_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```

```python
temp_df
```

```python
temp_df['is_duplicate'] = new_df['is_duplicate']
```

```python
temp_df.head()
```

# Train the Datset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(temp_df.iloc[:,0:-1].values,temp_df.iloc[:,-1].values,test_size=0.2,random_state=1)
```

# Applying RandomForestClassifier Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

# Applying XGBoost Classifier Model

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
accuracy_score(y_test,y_pred)
```

# Feature Engineering

```python
# Metrics Calculation
#
# q1_len    -> Character length of q1
# q2_len    -> Character length of q2
# q1_words  -> Number of words in q1
# q2_words  -> Number of words in q2
# words_common -> Number of unique words common to both q1 and q2
# words_total -> Total number of words in q1 + Total number of words in q2
# words_share -> Ratio of common words to the total number of words
#
# Explanation:
# - `q1_len` and `q2_len` provide the length of the questions in terms of characters.
# - `q1_words` and `q2_words` count the number of words in each question.
# - `words_common` calculates how many unique words are present in both questions.
# - `words_total` sums up the total number of words from both questions.
# - `words_share` is the ratio of `words_common` to `words_total`, representing the proportion of shared words.
```
